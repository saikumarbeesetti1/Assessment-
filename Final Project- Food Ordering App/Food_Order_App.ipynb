````````````````````````python
import json
import datetime

def register_user(user_json, name, paswrd, age, phn_no, email_id, adrs):
    user = {
        "id":1,
        "name": name,
        "Password": paswrd,
        "Age": age,
        "Phone number": phn_no,
        "Email ID": email_id,
        "Address": adrs,
        "Order history": {}
    }
    try:
        file = open("user.json", "r+")
        content = json.load(file)
        for i in range(len(content)):
            if content[i]["Phone number"] == phn_no:
                file.close()
                return "User alreaady exist"
        else:
            user["id"] = len(content) + 1
            content.append(user) 
    except JSONDecodeError:
        content = []
        content.append(user)
    file.seek(0)
    file.truncate()
    json.dump(content, file, indent=4)
    file.close()
    return "success"

def update_profile(user_json,user_id, paswrd, new_paswrd, phn_no, new_phn_no, email_id, new_email_id, adrs, new_adrs):
    file = open("user.json", "r+")
    content = json.load(file)
    for i in range(len(content)):
        if content[i]["id"] == user_id:
            if content[i]["Password"] == paswrd:
                content[i]["Password"] = new_paswrd
            if content[i]["Phone number"] == phn_no:
                content[i]["Phone number"] = new_phn_no
            if content[i]["Email ID"] == email_id:
                content[i]["Email ID"] = new_email_id
            if content[i]["Address"] == adrs:
                content[i]["Address"] = new_adrs
                print("```````````````````````")
                print("Your profile is updated")
                break
    file.seek(0)
    file.truncate()            
    json.dump(content, file, indent=4)
    file.close()
    return "Success"

def user_order_history(user_json, user_id):
    file = open("user.json", "r+")
    content = json.load(file)
    for i in range(len(content)):
        if content[i]["id"] == user_id:
            print("Order history")
            print("Date   |   Order")
            for i,j in content[i]["Order history"].items():
                print(f"{i} | {j}")
            file.close()
            return True
    file.close()
    return False

def place_order(user_json, food_json, user_id, food_name, quantity):
    date = datetime.datetime.today().strftime('%m-%d-%Y')
    file = open("user.json", "r+")
    content = json.load(file)
    file1 = open("food.json", "r+")
    content1 = json.load(file1)
    for i in range(len(content1)):
        if content1[i]["name"] == food_name:
            if content1[i]["no_of_plates"] >= quantity:
                for j in range(len(content)):
                    if content[j]["id"] == user_id:
                        content1[i]["no_of_plates"] -= quantity
                        if date not in content[j]["Order history"]:
                            content[j]["Order history"][date] = [content1[i]["name"]]
                        else:
                             content[j]["Order history"][date].append(content1[i]["name"])                                                               
            else:
                print("Pls enter less quantity")
                break
    
    file.seek(0)
    file.truncate()
    json.dump(content, file, indent=4)
    file.close()
    
    file1.seek(0)
    file1.truncate()
    json.dump(content1, file1, indent=4)
    file1.close()
    
def Add_food(food_json, food_name, no_of_plates, price, dscnt):
    food = {
        "id":1,
        "name": food_name,
        "no_of_plates": no_of_plates,
        "Price": price,
        "Discount":dscnt
    }
    try:
        fp = open("food.json", "r+")
        content = json.load(fp)
        for i in range(len(content)):
            if content[i]["name"] == food_name:
                fp.close()
                return "Food Is Already Available"
        food["id"] = len(content) + 1
        content.append(food)
    except JSONDecodeError:
        content = []
        content.append(food)
    fp.seek(0)
    fp.truncate()
    json.dump(content, fp, indent=4)
    fp.close()
    return "success"

def update_food(food_json, food_id, no_of_plates):
    file = open('food.json', "r+")
    content = json.load(file)
    for i in range(len(content)):
        if content[i]["id"] == food_id:
            content[i]["no_of_plates"] += no_of_plates
            print("*******************************")
            print(f"Food ID is: {content[i]['id']}")
            print(f"Total number of plates available are: {content[i]['no_of_plates']}")
            break
    file.seek(0)
    file.truncate()
    json.dump(content, file, indent=4)
    file.close()
    return"success"

def Edit_less_quantity_available(food_json, food_id, no_of_plates):
    file = open('food.json', "r+")
    content = json.load(file)
    for i in range(len(content)):
        if content[i]["id"] == food_id:
            content[i]["no_of_plates"] -= no_of_plates
            print("*******************************")
            print(f"Food ID is: {content[i]['id']}")
            print(f"Total number of plates available are: {content[i]['no_of_plates']}")
            break
    file.seek(0)
    file.truncate()
    json.dump(content, file, indent=4)
    file.close()
    return"success"

def remove_food(food_json, food_id):
    file = open("food.json", "r+")
    content = json.load(file)
    for i in range(len(content)):
        if content[i]["id"] == food_id:
            del content[i]
            file.seek(0)
            file.truncate()
            json.dump(content, file, indent=4)
            file.close
            return "success"
    return "Pls Enter Valid ID"

def View_list_of_food(food_json):
    file = open("food.json")
    content = json.load(file)
    print("-----------------------------------------------")
    print("       DHENU ATITYA HOTEL AND RESTAURANT       ")
    print("-----------------------------------------------")
    print("        ***MENU***       ")
    print("--------------------------")
    for i in range(len(content)):
        print("ID: ", content[i]["id"])
        print("----------------------")
        print(f"1. Name: {content[i]['name']}")
        print(f"2. NO_of_plates: {content[i]['no_of_plates']}")
        print(f"3. Price: {content[i]['Price']}")
        print(f"4. Discount: {content[i]['Discount']}")
    file.close()
    return True

val = input("Do you want to order food Y/N: ")
while val.lower() == "y":
    print("        ***HOME PAGE***        ")
    print("1) Register")
    print("2) Login")
    print("3) Exit")
    print("-------------------------------------------------")
    val1 = input("Choose one option from the above options: ")
    if val1 == "1":
        print()
        name = input("Enter your full name: ")
        paswrd = input("Create a password: ")
        age = int(input("Enter your age: "))
        phn_no = input("Enter your mobile number: ")
        email_id = input("Enter your email ID: ")
        adrs = input("Enter your address: ")
        register_user("user.json", name, paswrd, age, phn_no, email_id, adrs)
        print("--------------------------")
        print("You have successfully registered!!")
    elif val1 == "2":
        print()
        while True:
            print("1) User")
            print("2) Admin")
            print("3) Exit")
            print("------------------------------------")
            val2 = input("Choose one option from the above options: ")
            if val2 == "1":
                print("------User-----")
                user = input("Enter your full name: ")
                password = input("Enter a password: ")
                file = open("user.json", "r+")
                content = json.load(file)
                for i in range(len(content)):
                    if content[i]["name"] == user:
                        if content[i]["Password"] == password:
                            while True:
                                print()
                                print("1) View menu")
                                print("2) place your new order: ")
                                print("3) Update your profile: ")
                                print("4) show history of order: ")
                                print("5) Exit")
                                print("-----------------------------")
                                val3 = input("Select a option from the above: ")
                                if  val3 == "1":
                                    View_list_of_food("food.json")
                                elif val3 == "2":
                                    user_id = int(input("Enetr your user ID: "))
                                    food_name = input("Enter a food name: ")
                                    quantity = int(input("Enter a quantity: "))
                                    place_order("user.json", "food.json", user_id, food_name, quantity)
                                elif val3 == "3":
                                    user_id = int(input("Enter your user ID: "))
                                    paswrd = input("Enter a old password: ")
                                    new_paswrd = input("Create a new password: ")
                                    phn_no = input("Enter a old phone number: ")
                                    new_phn_no = input("Enter a new phone number: ")
                                    email_id = input("Enter a old email ID: ")
                                    new_email_id = input("Enter a new email ID: ")
                                    adrs = input("Enter your old address: ")
                                    new_adrs = input("Entre your new address: ")
                                    update_profile("user.json",user_id, paswrd, new_paswrd, phn_no, new_phn_no, email_id, new_email_id, adrs, new_adrs)
                                elif val3 == "4":
                                    user_id = int(input("Enter your user ID: "))
                                    user_order_history("user.json", user_id)
                                else:
                                    print("*********************")
                                    print("Thanks for your visit")
                                    break
            elif val2 == "2":
                print("+--+--Admin--+--+")
                user = input("Enter your name: ")
                password = input("Enter your password: ")
                file = open("admin.json", "r+")
                content = json.load(file)
                for i in range(len(content)):
                    if content[i]["name"] == user:
                        if content[i]["password"] == password:
                            while True:
                                print()
                                print("1) Add New Food")
                                print("2) Edit food")
                                print("3) View Food List")
                                print("4) Remove Food using food ID")
                                print("5) Edit food ('If less number of food is available')")
                                print("6) Exit!!")
                                print("-------------------------------------------------------")
                                val3 = input("Hello Admin!!, Please choose any option from the above list: ")
                                if val3 == "1":
                                    food_name = input("Enter a food name: ")
                                    no_of_plates = int(input("Enter the stock value: "))
                                    price = float(input("Enter price: "))
                                    dscnt = input("Enter discount: ")
                                    Add_food("food.json", food_name, no_of_plates, price, dscnt)
                                    print("---**---**---**---**---")
                                    print("New Food Is Added!!")
                                elif val3 == "2":
                                    food_id = int(input("Enter food ID: "))
                                    no_of_plates = int(input("Enter number of plates: "))
                                    update_food("food.json", food_id, no_of_plates)
                                    print("***************")
                                    print("Food Is Updated")
                                elif val3 == "3":
                                    View_list_of_food("food.json")
                                elif val3 == "4":
                                    food_id = int(input("Enter food ID: "))
                                    remove_food("food.json", food_id)
                                    print("---**---**---**---**---")
                                    print("Food Is Removed Successfully")
                                elif val3 == "5":
                                    food_id = int(input("Enter food ID: "))
                                    no_of_plates = int(input("Enter number of plates: "))
                                    Edit_less_quantity_available("food.json", food_id, no_of_plates)
                                    print("---**---**---**---**---")
                                    print("The Food has been edited")
                                else:
                                    print("Admin Exiting!!")
                                    break
                        else:
                            print("Wrong User Name :(")
                    else:
                        print("Wrong password :(")
            else:
                break
    else:
        print("Visit Again!!")
        
````````````````````````

```python
4. Filter the dataframe with only this columns
 ['continent','location','date','total_cases','total_deaths','gdp_per_capita','human_development_index'] and update the data frame.
```

```python
df1=df.filter(items=['continent','location','date','total_cases','total_deaths','gdp_per_capita','human_development_index'])
df.update(df1)
df.head()
```

```python
5. Data Cleaning

a. Remove all duplicates observations
b. Find missing values in all columns
c. Remove all observations where continent column value is  missing
Tip : using subset parameter in dropna
d. Fill all missing values with 0
```

```python
a. Remove all duplicates observations
```

```python
df=df.drop_duplicates()
df
```

```python
b. Find missing values in all columns
```

```python
print(" Please find below data, missing values will be shown as 'False' ")
print(" ")
print(df.isnull())
print(" ")
print(" ")

print(" Please find below data to know number of missing values each coulumns has ")
print(" ")
print(df.isnull().sum())
```

```python
c. Remove all observations where continent column value is missing
Tip: Using subset Parameter in dropna
```

```python
df.dropna(subset = ['continent'],inplace=True)
df['continent']
```

```python
d. Fill all missing values with 0
```

```python
df.fillna(0, inplace=True)
df
```

```python
6. Date time format:
    a. Convert date column in datetime format using pandas.to_datetime
    b. Create new column month after extracting month data from date column
```

```python
a. Convert date column in datetime format using pandas.to_datetime
```

```python
df['date']=pd.to_datetime(df['date']).dt.strftime('%m/%d/%Y') # here i changed date format to month/date/year
df
```

```python
b. Create new column month after extracting month data from date column
```

```python
df['month'] = pd.DatetimeIndex(df['date']).month
df
```

```python
7. Data Agregation:
    a. Find max value in all columns using groupby function on 'continent'
    Tip: use reset _index() after applying groupby
    b. Store the result in a new dataframe named 'df_groupby'. (Use df_groupby dataframe for all further analysis)
        
```

```python
a. Find max value in all columns using groupby function on 'continent'
    Tip: use reset _index() after applying groupby
```

```python
df.groupby('continent').max().reset_index()
```

```python
b. Store the result in a new dataframe named 'df_groupby'. (Use df_groupby dataframe for all further analysis)
```

```python
df_groupby = df.groupby('continent').max().reset_index()
df_groupby.head()
```

```python
8. Feature Engineering:
    a. Create a new feature 'total_deaths_to_total_cases' by ratio of 'total_deaths' column to 'total_cases'
```

```python
df['total_deaths_to_total_cases'] = df['total_deaths']/df['total_cases']
df['total_deaths_to_total_cases']
```

```python
b. Plot a scatter plot of 'total_cases' & 'gdp_per_capita'
```

```python
plt_scatter = plt.scatter(y=df['total_cases'],x=df['gdp_per_capita'],c ="orange")
plt.xlabel('gdp_per_capita')  
plt.ylabel('total_cases')  

plt_scatter.figure.savefig("plt_scatter.pdf")  # saving scatter plot in my drive

plt.show(plt_scatter)
```
